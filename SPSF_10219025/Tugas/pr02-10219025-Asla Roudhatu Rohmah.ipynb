Sebuah partikel bermuatan q bergerak dengan kecepatan $v(t)=v_{\rm x} (t)i+v_{\rm y}(t)j$ dalam ruang bermedan magnetik konstan 

$B=-kB_{\rm z}$

Tentukan gerak partikel :


## a. Tuliskan hukum newtonnya!

Hukum kedua Newton :

$$\sum F=ma\tag{1}$$

Gaya Lorentz :

$$F_{\rm B} = q (v x B)\tag{2}$$

$$F_{\rm B} = q ((v_{\rm x} (t)i + v_{\rm y}(t)j)x(-kB_{\rm z})\tag{3}$$

$$F_{\rm B} = q (-v_{\rm y}(t)B_{\rm z}i + v_{\rm x}(t)B_{\rm z}j)\tag{4}$$

$$F_{\rm B} = ma\tag{5}$$

$$q(-v_{\rm y}(t)B_{\rm z}i + v_{\rm x}(t)B_{\rm z})j =  ma\tag{6}$$

maka hukum newtonnya sebagai berikut : 

$$ q(-v_{\rm y}(t)B_{\rm z}i + v_{\rm x}(t)B_{\rm z})j = m(\frac{dv}{dt})  \tag{7}$$

dengan $$(\frac{dv}{dt})=(\frac{q(-v_{\rm y}(t)B_{\rm z}i + v_{\rm x}(t)B_{\rm z})j}{m})\tag{8}$$

## b. Tuliskan persamaan differensial terkopel antara kecepatan pada kedua arah!

$$(\frac{dv}{dt})=(\frac{q(-v_{\rm y}(t)B_{\rm z}i + v_{\rm x}(t)B_{\rm z})j}{m})\tag{9}$$

$$[(\frac{d}{dt})(v_{\rm x}(t)+v_{\rm y}(t))]+[(\frac{q}{m})(-B_{\rm z)}] x [v_{\rm x} (t)+v_{\rm y}(t))]=0\tag{10}$$

$$(\frac{dv_{\rm x}(t)}{dt})i+(\frac{dv_{\rm y}(t)}{dt})j+(\frac{qv_{\rm y}(t)B_{\rm z}}{m})i-(\frac{qv_{\rm X}(t)B_{\rm z}}{m})j=0\tag{11}$$

$$[(\frac{dv_{\rm x}(t)}{dt})+(\frac{qv_{\rm y}(t)B_{\rm z}}{m})i]+[(\frac{dv_{\rm y}(t)}{dt})-(\frac{qv_{\rm X}(t)B_{\rm z}}{m})j]=0\tag{12}$$

asumsikan $$v_{\rm x} = v_{\rm x} (t)\tag{13}$$ dan $$v_{\rm y} = v_{\rm y} (t)\tag{14}$$

maka akan menghasilkan persamaan differensial terkopel antara kecepatan pada dua komponen arah : 
1. $$(\frac{dv_{\rm x}(t)}{dt})+(\frac{qB_{\rm z}}{m})(v_{\rm y}(t))=0\tag{15}$$
2. $$(\frac{dv_{\rm y}(t)}{dt})-(\frac{q(t)B_{\rm z}}{m})(v_{\rm x}(t))=0\tag{16}$$

## c. Selesaikan kedua persamaan differensial sehingga dapat diperoleh $v_{\rm x}(t), v_{\rm y}(t), x(t),$ dan $y(t)$. Lakukan sesuai teori!

misalkan $$w=(\frac{qB_{\rm z}}{m})\tag{17}$$

maka $$v_{\rm x}=(\frac{1}{w})(\frac{dv_{\rm y}}{dt})$$ dan $$v_{\rm y}=-(\frac{1}{w})(\frac{dv_{\rm x}}{dt})\tag{18}$$

Persamaan Differensial Linier Orde Dua

$$v_{\rm x}=-(\frac{1}{w^2})(\frac{d^2v_{\rm y}}{dt^2})$$ dan $$v_{\rm y}=-(\frac{1}{w^2})(\frac{d^2v_{\rm x}}{dt^2})\tag{19}$$

Persamaan Umum

$$v_{\rm x,y}=A cos wt+B sin wt\tag{20}$$

$$v_{\rm x}=u cos wt\tag{21}$$

$$v_{\rm y}=u  sin wt\tag{22}$$

Jika $$t=t_{\rm 0}\tag{23}$$ maka $$x(t_{\rm 0})=x_{\rm 0}\tag{24}$$ dan $$y(t_{\rm 0})=y_{\rm 0}\tag{25}$$

$$v_{\rm x}=u cos [w(t-t_{\rm 0})+\phi_{\rm 0}]\tag{21}$$

$$v_{\rm y}=u  sin [w(t-t_{\rm 0})+\phi_{\rm 0}]\tag{22}$$

Persamaan 21 dan 22 diintegralkan menjadi 

$$x(t)=(\frac{u}{w})sin [w(t-t_{\rm 0})+\phi_{\rm 0}]\tag{23}$$

$$y(t)=-(\frac{u}{w})cos [w(t-t_{\rm 0})+\phi_{\rm 0}]\tag{24}$$

Dapat diuji secara numerik untuk menentukan gambaran gerak rotasi partikel berdasarkan perhitungan

```python
import numpy as np
import matplotlib.pyplot as plt

t=np.arange(0,20,0.01)

vx=20*np.cos(t)
vy=-20*np.sin(t)

plt.figure()
plt.plot(t,vx,'g', label = 'vx')
plt.plot(t,vy,'y', label = 'vy')
plt.title('Gerak Rotasi Partikel')
plt.xlabel('Waktu')
plt.ylabel('Kecepatan')
plt.legend(loc=0)
plt.grid(True)

plt.show()
```

## d. Tentukan solusi numeriknya!


Persamaan (15) dan (16) dimodifikasi menjadi

$$(\frac{dv_{\rm x}(t)}{dt})=-(\frac{qB_{\rm z}}{m})(v_{\rm y}(t))\tag{25}$$

$$(\frac{dv_{\rm y}(t)}{dt})=(\frac{q(t)B_{\rm z}}{m})(v_{\rm x}(t))\tag{26}$$

```python
#Mendefinisikan persamaan Runge Kutta Orde-4
def run_kut4(F,t,x,y,h):
        k0=h*F(t,x,y)
        k1=h*F(t+h/2.0, x+k0/2.0,y)
        k2=h*F(t+h/2.0,x+k1/2.0,y)
        k3=h*F(t+h,x+k2,y)
        return (k0+2.0*k1+2.0*k2+k3)/6.0

def vx_partikel(t,vx,vy,q=1, B=1,m=1): #Persamaan 25
    return -q*B/m*vy
def vy_partikel(t,vy,vx,q=1, B=1,m=1): #Persamaan 26
    return q*B/m*vx

    
#Initial Condition
vx_awal=0
vy_awal=20
t_awal=0
t_akhir=20
h=0.01
vx_s=[vx_awal] ; vy_s=[vy_awal] ; t_s=[t_awal]
while t_s[-1]<t_akhir:
    vx=vx_s[-1]+run_kut4(vx_partikel,t_s[-1],vx_s[-1],vy_s[-1],h)
    vy=vy_s[-1]+run_kut4(vy_partikel,t_s[-1],vy_s[-1],vx_s[-1],h)
    h=min(h,t_akhir-t_s[-1])
    t=t_s[-1]+h
    
    t_s.append (t)
    vx_s.append(vx)
    vy_s.append(vy)

#Plotting
plt.figure()
plt.plot(t_s,vy_s,t_s,vx_s)
plt.plot(t_s,vx_s,'y', label = 'vx')
plt.plot(t_s,vy_s,'g', label = 'vy')
plt.title('Gerak Rotasi Partikel')
plt.title ('Rotasi partikel RK-4')
plt.xlabel('Waktu')
plt.ylabel('Kecepatan')
plt.legend(loc=0)
plt.grid(True)

plt.show()
```
